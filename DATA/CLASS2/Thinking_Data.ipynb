```python
import json
```

# 1) load the data

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data    
```

```python
data = load_data("Store_data.json")
print(data,type(data))
```

# 2) clean and structure data

```python
def clean_data(data):

    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data = []
    unique_users = set()
    
    for user in data:
         # clean ratings (convert to string safely)
        raw_rating = str(user["rating"]).strip().lower()   
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        
        # handle age missing
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        
         # deduplication by user name (removing extra spaces)
        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data   # <-- corrected indentation
```

```python
data = clean_data(data)
clean_data(data)
```

# 3) get meaningful insights from data

```python
def get_insights(data):

    # avg rating

    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])
        
    print(f"average rating = {tot_rating/len(data)}")  



    # percentage of users with poor rating

    poor_ratings = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_ratings += 1

    print(f"% of user with poor rating = {poor_ratings/len(data) * 100}%")        
```

```python
get_insights(data)
```

# 4) Recommendation feature

```python
def get_recommendations(data):
    recommendations = []

    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]
        
        if(float(user["rating"])>=4):
           curr_recomm["brand"] = "apple"
        else:
           curr_recomm["brand"] = "samsung"

        recommendations.append(curr_recomm)
    return recommendations   
```

```python
get_recommendations(data)
```
